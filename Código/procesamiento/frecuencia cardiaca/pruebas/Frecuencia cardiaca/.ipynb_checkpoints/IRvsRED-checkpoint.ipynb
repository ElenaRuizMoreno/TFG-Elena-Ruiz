# Procesamiento de datos y estimación HR 

### Análisis de la señal IR

#### 1.Cargar y visualizar los datos
Primero, lee el archivo CSV en un DataFrame de Pandas y visualiza la señal IR en función del tiempo.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np

# Cargar datos
file_path = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos crudos\save_log2\raw_data_97_119.csv"
df = pd.read_csv(file_path, sep=";")

# Convertir tiempo a segundos
df["Tiempo (s)"] = df["Tiempo (ms)"] / 1000.0

# Graficar señales IR y RED
plt.figure(figsize=(10, 4))
plt.plot(df["Tiempo (s)"], df["IR"], label="Señal IR", color="red")
plt.plot(df["Tiempo (s)"], df["RED"], label="Señal RED", color="blue", alpha=0.7)
plt.xlabel("Tiempo (s)")
plt.ylabel("Intensidad")
plt.title("Comparación de Señales IR y RED")
plt.legend()
plt.show()
```

#### 2. Identificar los picos de la señal IR
La frecuencia cardíaca se calcula a partir del número de picos en la señal IR. Para detectar los picos, puedes usar scipy.signal.find_peaks.

```python
from scipy.signal import find_peaks

# Encontrar picos en la señal IR
peaks, _ = find_peaks(df["IR"], distance=50)  # Ajustar el parámetro 'distance' según la frecuencia esperada

# Graficar señal con picos detectados
plt.figure(figsize=(10, 4))
plt.plot(df["Tiempo (s)"], df["IR"], label="Señal IR", color="red")
plt.plot(df["Tiempo (s)"][peaks], df["IR"][peaks], "bo", label="Picos detectados")  # Marcar picos
plt.xlabel("Tiempo (s)")
plt.ylabel("Intensidad IR")
plt.title("Detección de Picos en la Señal IR")
plt.legend()
plt.show()
```

#### 3. Calcular la frecuencia cardíaca
La HR se obtiene midiendo el tiempo entre picos sucesivos y convirtiéndolo a latidos por minuto (BPM).

```python
import numpy as np

# Calcular diferencias de tiempo entre picos consecutivos
peak_times = df["Tiempo (s)"][peaks].values
rr_intervals = np.diff(peak_times)  # Intervalos RR en segundos

# Calcular frecuencia cardíaca (HR) en BPM
hr_bpm = 60 / rr_intervals  # Conversión a latidos por minuto

# Mostrar resultado
print(f"Frecuencia cardíaca media: {np.mean(hr_bpm):.2f} BPM")
```

La señal IR en muchos pulsioxímetros la frecuencia cardíaca se extrae principalmente de esa señal, ya que suele tener una mejor relación señal-ruido para detectar los latidos. Pero, en realidad, podemos analizar también la señal RED y ver cuál es más útil o si ambas aportan información complementaria.

- **IR suele ser más estable:** La absorción del infrarrojo en el tejido cambia con el pulso sanguíneo y suele ser menos afectada por interferencias externas.
- **RED también varía con el pulso:** Puede ser útil para detectar los latidos y, además, es fundamental para calcular la saturación de oxígeno (SpO₂).
- **Comparación entre ambas señales:** Analizar ambas señales nos puede ayudar a mejorar la precisión y evitar errores debidos a ruido o interferencias.

### Comparación señales IR con señales de luz roja

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np

# Cargar datos
file_path = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos crudos\save_log2\raw_data_97_119.csv"
df = pd.read_csv(file_path, sep=";")

# Convertir tiempo a segundos
df["Tiempo (s)"] = df["Tiempo (ms)"] / 1000.0

# Graficar señales IR y RED
plt.figure(figsize=(10, 4))
plt.plot(df["Tiempo (s)"], df["IR"], label="Señal IR", color="red")
plt.plot(df["Tiempo (s)"], df["RED"], label="Señal RED", color="blue", alpha=0.7)
plt.xlabel("Tiempo (s)")
plt.ylabel("Intensidad")
plt.title("Comparación de Señales IR y RED")
plt.legend()
plt.show()
```

```python
# Detectar picos en IR y RED
peaks_ir, _ = find_peaks(df["IR"], distance=50)
peaks_red, _ = find_peaks(df["RED"], distance=50)

# Graficar detección de picos
plt.figure(figsize=(10, 4))
plt.plot(df["Tiempo (s)"], df["IR"], label="Señal IR", color="red")
plt.plot(df["Tiempo (s)"], df["RED"], label="Señal RED", color="blue", alpha=0.7)
plt.plot(df["Tiempo (s)"][peaks_ir], df["IR"][peaks_ir], "ro", label="Picos IR")
plt.plot(df["Tiempo (s)"][peaks_red], df["RED"][peaks_red], "bo", label="Picos RED")
plt.xlabel("Tiempo (s)")
plt.ylabel("Intensidad")
plt.title("Detección de Picos en IR y RED")
plt.legend()
plt.show()
```

```python
# Calcular intervalos RR y HR en BPM para IR
peak_times_ir = df["Tiempo (s)"][peaks_ir].values
rr_intervals_ir = np.diff(peak_times_ir)
hr_bpm_ir = 60 / rr_intervals_ir

# Calcular intervalos RR y HR en BPM para RED
peak_times_red = df["Tiempo (s)"][peaks_red].values
rr_intervals_red = np.diff(peak_times_red)
hr_bpm_red = 60 / rr_intervals_red

# Mostrar resultados
print(f"Frecuencia cardíaca media IR: {np.mean(hr_bpm_ir):.2f} BPM")
print(f"Frecuencia cardíaca media RED: {np.mean(hr_bpm_red):.2f} BPM")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np
import os

# Ruta de la carpeta con los CSV
folder_path = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos crudos\save_log2"

# Obtener la lista de archivos CSV en la carpeta
csv_files = [f for f in os.listdir(folder_path) if f.endswith(".csv") and f != "HR_results.csv"]

# Crear listas para almacenar resultados
hr_results = []

# Recorrer cada archivo CSV
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    print(f"\n🔹 Analizando: {file}")

    # Cargar datos
    df = pd.read_csv(file_path, sep=";")

    # Convertir tiempo a segundos
    df["Tiempo (s)"] = df["Tiempo (ms)"] / 1000.0

    # Detectar picos en IR y RED
    peaks_ir, _ = find_peaks(df["IR"], distance=50)
    peaks_red, _ = find_peaks(df["RED"], distance=50)

    # Calcular HR en BPM para IR
    if len(peaks_ir) > 1:
        rr_intervals_ir = np.diff(df["Tiempo (s)"][peaks_ir].values)
        hr_bpm_ir = 60 / rr_intervals_ir
        hr_mean_ir = np.mean(hr_bpm_ir)
    else:
        hr_mean_ir = np.nan

    # Calcular HR en BPM para RED
    if len(peaks_red) > 1:
        rr_intervals_red = np.diff(df["Tiempo (s)"][peaks_red].values)
        hr_bpm_red = 60 / rr_intervals_red
        hr_mean_red = np.mean(hr_bpm_red)
    else:
        hr_mean_red = np.nan

    # Guardar resultados
    hr_results.append([file, hr_mean_ir, hr_mean_red])

    # Graficar la señal con los picos detectados
    plt.figure(figsize=(10, 4))
    plt.plot(df["Tiempo (s)"], df["IR"], label="Señal IR", color="red")
    plt.plot(df["Tiempo (s)"], df["RED"], label="Señal RED", color="blue", alpha=0.7)
    plt.plot(df["Tiempo (s)"][peaks_ir], df["IR"][peaks_ir], "ro", label="Picos IR")
    plt.plot(df["Tiempo (s)"][peaks_red], df["RED"][peaks_red], "bo", label="Picos RED")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Intensidad")
    plt.title(f"Detección de Picos - {file}")
    plt.legend()
    plt.show()

# Crear un DataFrame con los resultados
df_results = pd.DataFrame(hr_results, columns=["Archivo", "HR Media IR (BPM)", "HR Media RED (BPM)"])

# Mostrar los resultados
print("\n📊 Resultados de Frecuencia Cardíaca:")
print(df_results)

# Guardar los resultados en un CSV
output_results = os.path.join(folder_path, "HR_results.csv")
df_results.to_csv(output_results, index=False)
print(f"\n✅ Resultados guardados en: {output_results}")
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, butter, filtfilt

# Ruta de la carpeta donde están los datos
folder_path = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos crudos\save_log2"

# Función para filtrar la señal y reducir ruido
def butter_lowpass_filter(data, cutoff=2.0, fs=100, order=5):
    nyq = 0.5 * fs  # Frecuencia de Nyquist
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

# Leer todos los archivos CSV en la carpeta
data_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

# Procesar cada archivo
for file in data_files:
    file_path = os.path.join(folder_path, file)
    print(f'🔹 Analizando: {file}')
    
    # Cargar datos
    df = pd.read_csv(file_path, delimiter=';', decimal='.')
    df["Tiempo (s)"] = df["Tiempo (ms)"] / 1000.0
    
    # Filtrar señal IR y RED
    df["IR_filtrado"] = butter_lowpass_filter(df["IR"], cutoff=2.0, fs=100)
    df["RED_filtrado"] = butter_lowpass_filter(df["RED"], cutoff=2.0, fs=100)
    
    # Detectar picos
    peaks_ir, _ = find_peaks(df["IR_filtrado"], distance=50)
    peaks_red, _ = find_peaks(df["RED_filtrado"], distance=50)
    
    # Calcular frecuencia cardiaca
    if len(peaks_ir) > 1:
        rr_intervals = np.diff(df["Tiempo (s)"][peaks_ir])
        hr_ir = 60 / np.mean(rr_intervals)
    else:
        hr_ir = np.nan
    
    if len(peaks_red) > 1:
        rr_intervals_red = np.diff(df["Tiempo (s)"][peaks_red])
        hr_red = 60 / np.mean(rr_intervals_red)
    else:
        hr_red = np.nan
    
    # Graficar señales y picos detectados
    plt.figure(figsize=(10, 5))
    plt.plot(df["Tiempo (s)"], df["IR_filtrado"], label="IR Filtrado", color='blue')
    plt.plot(df["Tiempo (s)"][peaks_ir], df["IR_filtrado"][peaks_ir], "ro", label="Picos IR")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Señal")
    plt.title(f"Detección de Picos - {file}")
    plt.legend()
    plt.show()
    
    print(f'📊 Frecuencia cardíaca estimada: {hr_ir:.2f} BPM (IR) - {hr_red:.2f} BPM (RED)')
```
