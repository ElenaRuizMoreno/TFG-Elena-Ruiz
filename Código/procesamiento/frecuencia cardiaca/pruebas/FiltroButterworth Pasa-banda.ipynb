## Filtro Butterworth pasa-banda.

- Elimina frecuencias fuera del rango de interés, dejando solo las que corresponden a la frecuencia cardíaca.
- Paso-banda de 0.5 Hz a 4 Hz, lo que equivale a 30 - 240 latidos por minuto (bpm).
- Suaviza la señal y elimina ruido de alta frecuencia (movimientos, interferencias) y baja frecuencia (tendencias lentas, como la variación de luz ambiental).


El filtro elimina las componentes de frecuencia no deseadas, y la salida es una oscilación en torno a 0 en lugar de estar alrededor del valor original de la señal. Esto es normal y significa que tenemos una versión limpia de la variabilidad pulsátil.

```python
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt

# Definir función para aplicar un filtro Butterworth pasa-banda
def butter_bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs  # Frecuencia de Nyquist
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

# Calcular frecuencia de muestreo en Hz (inversa del intervalo medio de tiempo)
fs = 1000 / df["Tiempo (ms)"].diff().mean()  

# Aplicar filtro pasa-banda (0.5 Hz - 4 Hz)
filtered_ir = butter_bandpass_filter(df["IR"], lowcut=0.5, highcut=4, fs=fs, order=4)

plt.figure(figsize=(12, 5))
plt.plot(df["Tiempo (ms)"], df["IR"], label="IR Original", alpha=0.5)
plt.plot(df["Tiempo (ms)"], filtered_ir, label="IR Filtrada", color="red")
plt.xlabel("Tiempo (ms)")
plt.ylabel("Intensidad IR")
plt.title("Señal IR - Original vs Filtrada")
plt.legend()
plt.grid(True)
plt.show()
```

## Detectar picos

```python
from scipy.signal import find_peaks
import numpy as np
import matplotlib.pyplot as plt

# Detectar picos en la señal filtrada
peaks, _ = find_peaks(filtered_ir, height=0, distance=fs*0.27)  # Distancia mínima de 0.5s entre picos (120 BPM)

# Graficar señal filtrada con picos detectados
plt.figure(figsize=(12, 5))
plt.plot(df["Tiempo (ms)"], filtered_ir, label="IR Filtrada", color="red")
plt.plot(df["Tiempo (ms)"].iloc[peaks], filtered_ir[peaks], "bo", label="Picos detectados")  # Marcar picos en azul
plt.xlabel("Tiempo (ms)")
plt.ylabel("Intensidad Filtrada")
plt.title("Detección de Picos en la Señal IR")
plt.legend()
plt.grid(True)
plt.show()

# Calcular la frecuencia cardíaca en BPM
if len(peaks) > 1:
    rr_intervals = np.diff(df["Tiempo (ms)"].iloc[peaks]) / 1000  # Diferencias en segundos
    heart_rate = 60 / rr_intervals.mean()  # BPM
    print(f"Frecuencia Cardíaca Estimada: {heart_rate:.2f} BPM")
else:
    print
```

## FFT (Transformada de Fourier)

En vez de contar picos, analizar las frecuencias de la señal y detectar la más predominante, que debería corresponder a la frecuencia cardíaca.

Este método convierte la señal al dominio de la frecuencia para detectar la frecuencia dominante

```python
from scipy.fft import fft, fftfreq

# Obtener la señal sin tendencia (restar la media)
ir_signal = filtered_ir - np.mean(filtered_ir)


# Aplicar FFT
N = len(ir_signal)
T = 1/fs  # Período de muestreo
yf = fft(ir_signal)
xf = fftfreq(N, T)  # Frecuencias asociadas

# Filtrar solo frecuencias positivas
xf = xf[:N // 2]
yf = np.abs(yf[:N // 2])

# Encontrar el pico más alto en el espectro
peak_freq = xf[np.argmax(yf)]  # Frecuencia en Hz
heart_rate_fft = peak_freq * 60  # Convertir a BPM

plt.figure(figsize=(10, 5))
plt.plot(xf, yf)
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.title("Transformada de Fourier de la señal IR")
plt.grid()
plt.show()

print(f"Frecuencia Cardíaca estimada por FFT: {heart_rate_fft:.2f} BPM")
```
