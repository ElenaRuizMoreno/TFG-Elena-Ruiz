# Código para hacer una limpieza de datos personalizada para cada señal

Este código tiene como objetivo visualizar cada señal completa, valorar su calidad y ver a partir de qué segundos podemos considerar la señal para poder realizar un buen análisis o flitrado y a parte, no perder información.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

ruta_entrada = r'C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos crudos\save_log2'
ruta_salida = r'C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos_limpios'
fs = 60  # Frecuencia de muestreo (Hz)

if not os.path.exists(ruta_salida):
    os.makedirs(ruta_salida)

for archivo in os.listdir(ruta_entrada):
    if archivo.endswith('.csv'):
        print(f'\nProcesando {archivo}')
        df = pd.read_csv(os.path.join(ruta_entrada, archivo), delimiter =";")

        # Señales útiles
        df['IR_net'] = df['IR'] - df['AMB_IR']
        df['RED_net'] = df['RED'] - df['AMB_RED']

        # Graficar
        plt.figure(figsize=(10,4))
        plt.plot(df['Tiempo (ms)'], df['IR_net'], label='IR')
        plt.plot(df['Tiempo (ms)'], df['RED_net'], label='RED')
        plt.legend()
        plt.title(f'{archivo}')
        plt.xlabel('Tiempo (s)')
        plt.show()

        # Input manual
        t_inicial = float(input('¿Desde qué segundo te quieres quedar?: '))
        t_final = float(input('¿Hasta qué segundo te quieres quedar?: '))

        # Filtrar
        df_limpio = df[(df['Tiempo (ms)'] >= t_inicial) & (df['Tiempo (ms)'] <= t_final)]

        # Guardar
        salida = os.path.join(ruta_salida, archivo)
        df_limpio.to_csv(salida, index=False)
        print(f'Guardado limpio en {salida}')
```
