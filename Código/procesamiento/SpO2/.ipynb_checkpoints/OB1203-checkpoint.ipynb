## 1. Cargar, restar luz ambiente, aplicar filtros

```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# 1. Cargar datos
def load_ppg_data(path):
    df = pd.read_csv(path)
    df.columns = ["Tiempo_ms", "IR", "AMB_IR", "RED", "AMB_RED"]
    return df

# 2. Restar señal ambiente
def subtract_ambient_signals(df):
    df["IR_corr"] = df["IR"] - df["AMB_IR"]
    df["RED_corr"] = df["RED"] - df["AMB_RED"]
    return df

# 3. Media móvil pasa bajo
def moving_average(signal, window_size=8):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='same')

# 4. Filtro pasa alto tipo paper (resta media centrada)
def highpass_filter_centered(signal, window_size=64):
    baseline = pd.Series(signal).rolling(window=window_size, center=True, min_periods=1).mean()
    return signal - baseline

# 5. Calcular AC y DC de una señal (por ejemplo, RED_corr)
def calculate_ac_dc(signal):
    dc = np.mean(signal)
    ac = np.mean(np.abs(signal - dc))  # desviación media
    return ac, dc
```

## 2. Calcular AC y DC
Según el paper, usamos:

- DC = media

- AC = desviación media respecto a cero (no RMS)

```python
# 5. Calcular AC y DC de una señal (por ejemplo, RED_corr)
def calculate_ac_dc(signal):
    dc = np.mean(signal)
    ac = np.mean(np.abs(signal - dc))  # desviación media
    return ac, dc
```

## 3. Estimar Ratio y SpO

```python
# 6. Calcular ratio R
def calculate_ratio(ac_red, dc_red, ac_ir, dc_ir):
    return (ac_red / dc_red) / (ac_ir / dc_ir)

# 7. Estimar SpO2 con fórmula lineal (c0 y c1 pueden calibrarse)
def estimate_spo2(ratio, c0=1.1, c1=0.25):
    return (c0 - c1 * ratio) * 100
```

```python
def process_spo2_pipeline(path_csv):
    df = load_ppg_data(path_csv)
    df = subtract_ambient_signals(df)

    # Aplicar media móvil
    df["IR_filtered"] = moving_average(df["IR_corr"])
    df["RED_filtered"] = moving_average(df["RED_corr"])

    # Aplicar filtro pasa alto
    df["IR_hp"] = highpass_filter_centered(df["IR_filtered"])
    df["RED_hp"] = highpass_filter_centered(df["RED_filtered"])

    # Recortamos los primeros segundos si quieres (ej. primeros 3 s)
    df = df[df["Tiempo_ms"] >= df["Tiempo_ms"].iloc[0] + 3000].reset_index(drop=True)

    # Calcular AC y DC con las señales finales
    ac_ir, dc_ir = calculate_ac_dc(df["IR_hp"])
    ac_red, dc_red = calculate_ac_dc(df["RED_hp"])

    ratio = calculate_ratio(ac_red, dc_red, ac_ir, dc_ir)
    spo2 = estimate_spo2(ratio)

    print(f"SpO₂ estimado: {spo2:.2f}%")

    return df, spo2
```

```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# --- FUNCIONES DEL PIPELINE ---

def load_ppg_data(path):
    df = pd.read_csv(path)
    df.columns = ["Tiempo_ms", "IR", "AMB_IR", "RED", "AMB_RED"]
    return df

def subtract_ambient_signals(df):
    df["IR_corr"] = df["IR"] - df["AMB_IR"]
    df["RED_corr"] = df["RED"] - df["AMB_RED"]
    return df

def moving_average(signal, window_size=8):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='same')

def highpass_filter_centered(signal, window_size=64):
    baseline = pd.Series(signal).rolling(window=window_size, center=True, min_periods=1).mean()
    return signal - baseline

def calculate_ac_dc(signal):
    dc = np.mean(signal)
    ac = np.mean(np.abs(signal - dc))  # desviación media respecto a la media
    return ac, dc

def calculate_ratio(ac_red, dc_red, ac_ir, dc_ir):
    return (ac_red / dc_red) / (ac_ir / dc_ir)

def estimate_spo2(ratio, c0=1.1, c1=0.25):
    return (c0 - c1 * ratio) * 100

# --- PROCESAMIENTO INDIVIDUAL ---

def process_spo2_pipeline(path_csv):
    df = load_ppg_data(path_csv)
    df = subtract_ambient_signals(df)
    df["IR_filtered"] = moving_average(df["IR_corr"])
    df["RED_filtered"] = moving_average(df["RED_corr"])
    df["IR_hp"] = highpass_filter_centered(df["IR_filtered"])
    df["RED_hp"] = highpass_filter_centered(df["RED_filtered"])
    df = df[df["Tiempo_ms"] >= df["Tiempo_ms"].iloc[0] + 3000].reset_index(drop=True)
    ac_ir, dc_ir = calculate_ac_dc(df["IR_hp"])
    ac_red, dc_red = calculate_ac_dc(df["RED_hp"])
    ratio = calculate_ratio(ac_red, dc_red, ac_ir, dc_ir)
    spo2 = estimate_spo2(ratio)
    return df, spo2, ratio

# --- PROCESAR TODOS LOS ARCHIVOS EN UNA CARPETA ---

def run_batch_pipeline(folder_path):
    results = []

    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            path = os.path.join(folder_path, file)

            # Extraer SpO2 real desde el nombre del archivo
            match = re.search(r'raw_data_(\d+)_\d+', file)
            spo2_real = int(match.group(1)) if match else None

            df, spo2_est, ratio = process_spo2_pipeline(path)
            results.append((file, spo2_est, spo2_real, ratio))

            # Plot
            plt.figure(figsize=(10, 4))
            plt.plot(df["Tiempo_ms"], df["IR_hp"], label='IR (filtrada)', color='black')
            plt.plot(df["Tiempo_ms"], df["RED_hp"], label='RED (filtrada)', color='red')
            plt.title(f"{file}\nSpO₂ estimado: {spo2_est:.1f}% | SpO₂ real: {spo2_real}%")
            plt.xlabel("Tiempo (ms)")
            plt.ylabel("Amplitud (a.u.)")
            plt.legend()
            plt.grid(True)
            plt.tight_layout()
            plt.show()

    return pd.DataFrame(results, columns=["Archivo", "SpO2 estimado", "SpO2 real", "Ratio R"])
```
