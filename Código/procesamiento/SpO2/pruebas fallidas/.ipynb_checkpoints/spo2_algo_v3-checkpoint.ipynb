## Estimación avanzada de SpO₂ basada en detección de latidos y cálculo adaptativo del ratio

Este notebook implementa una tercera versión mejorada del algoritmo para estimar la saturación de oxígeno en sangre (SpO2) a partir de señales PPG (IR y RED). El enfoque combina preprocesamiento de señal, detección robusta de latidos y cálculo de la SpO2 a partir de la relación AC/DC en cada latido.

### Procesamiento aplicado

1. **Carga y preprocesamiento de datos**
   - Lectura de archivos CSV con señales crudas de pulsioxímetro (IR, RED y ambiente).
   - Corrección de señal restando la componente de luz ambiente (AMB_IR y AMB_RED).
   - Recorte de los primeros segundos para evitar artefactos de inicio.

2. **Filtrado de señales**
   - Aplicación de un filtro paso bajo o suavizado para limpiar ruido de alta frecuencia y estabilizar la forma de onda.

3. **Detección de picos (latidos)**
   - Identificación de valles o picos en la señal IR para segmentar la señal en ciclos cardíacos.

4. **Cálculo por latido**
   - Para cada ciclo, se calcula la amplitud AC y el nivel DC de las señales IR y RED.
   - Se computa el ratio R y se estima la SpO2 con una fórmula empírica.

5. **Evaluación y visualización**
   - Se imprime el valor medio de SpO2 estimado por archivo.
   - Se puede representar la evolución de la SpO2 estimada a lo largo del tiempo o por ventana.

### Tabla de referencia spo2_table
Define la tabla de valores de SpO2 que se usa normalmente con el pulsioxímetro MAX30102.

Esta tabla contiene valores de saturación de oxígeno (SpO2) que se usan como referencia.
La idea es que se va a calcular un ratio (R) a partir de las señales IR y RED, y con ese ratio buscar el valor correspondiente de SpO2 en esta tabla.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# Ruta CSV
folder_path = r'C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos_limpios'

# Tabla de SpO2
spo2_table = np.array([
    95, 95, 95, 96, 96, 96, 97, 97, 97, 97, 97, 98, 98, 98, 98, 98, 99, 99, 99, 99,
    99, 99, 99, 99, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100, 100,
    100, 100, 100, 100, 99, 99, 99, 99, 99, 99, 99, 99, 98, 98, 98, 98, 98, 98, 97, 97,
    97, 97, 96, 96, 96, 96, 95, 95, 95, 94, 94, 94, 93, 93, 93, 92, 92, 92, 91, 91,
    90, 90, 89, 89, 89, 88, 88, 87, 87, 86, 86, 85, 85, 84, 84, 83, 82, 82, 81, 81,
    80, 80, 79, 78, 78, 77, 76, 76, 75, 74, 74, 73, 72, 72, 71, 70, 69, 69, 68, 67,
    66, 66, 65, 64, 63, 62, 62, 61, 60, 59, 58, 57, 56, 56, 55, 54, 53, 52, 51, 50,
    49, 48, 47, 46, 45, 44, 43, 42, 41, 40, 39, 38, 37, 36, 35, 34, 33, 32
])

def calculate_spo2(ir_data, red_data):
    ir_ac = np.max(ir_data) - np.min(ir_data)
    ir_dc = np.mean(ir_data)

    red_ac = np.max(red_data) - np.min(red_data)
    red_dc = np.mean(red_data)

    if ir_ac == 0 or red_ac == 0 or ir_dc == 0 or red_dc == 0:
        return 0  # señal no válida

    ratio = (red_ac * ir_dc) / (ir_ac * red_dc) * 100
    ratio = int(ratio)

    # Limitar el ratio al rango de la tabla
    ratio = min(max(ratio, 0), len(spo2_table) - 1)

    spo2 = spo2_table[ratio]
    return spo2

# Leer todos los CSV de la carpeta
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        ir_signal = df['IR'].values
        red_signal = df['RED'].values

        spo2 = calculate_spo2(ir_signal, red_signal)

        print(f'Archivo: {filename} -> SpO2 estimado: {spo2}%')
        
        # Si quieres ver las señales:
        plt.figure(figsize=(12, 6))
        plt.plot(ir_signal, label='IR')
        plt.plot(red_signal, label='RED')
        plt.legend()
        plt.title(f'{filename} - SpO2: {spo2}%')
        plt.show()

```
