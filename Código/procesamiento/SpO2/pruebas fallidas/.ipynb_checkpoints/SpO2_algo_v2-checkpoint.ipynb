# Cálculo de la SpO2 a partir de los crudos del sensor

Este código implementa la lógica que transforma señales IR y RED en una estimación de SpO₂ y HR. El proceso es más o menos así:

1. **Preprocesamiento de la señal IR**
- Se calcula la media (DC component) de la señal IR.

- Se elimina esa media para quedarte solo con la parte AC (las oscilaciones que indican los latidos).

- Se invierte la señal para facilitar la detección de valles (latidos).

2. **Filtrado con media móvil**
Se aplica una media móvil de 4 puntos para suavizar la señal.

3. **Detección de valles (latidos)**
- Se usa un detector de picos sobre la señal invertida (porque los valles son picos tras invertir).

- Se calcula la frecuencia cardíaca a partir del tiempo entre estos valles.

4. **Preparación para cálculo de SpO2**
- Se vuelve a cargar la señal IR y RED sin invertir ni filtrar.

- Se localizan las posiciones de los valles detectados en IR.

- Para cada intervalo entre dos valles, se calcula:

    - AC (variación entre los extremos y el máximo del pulso)

    - DC (valor máximo del pulso)

- Y se calcula el ratio:

    R = ((AC_RED/DC_RED)/(AC_IR/DC_IR))
 
    Este ratio se convierte en SpO2 usando una tabla o fórmula empírica (en este caso, uch_spo2_table o una fórmula cuadrática comentada al final).

5. **Postprocesado**
- Se toman los 5 valores de ratio más representativos.

- Se calcula la mediana para evitar errores por latidos anómalos.

- Si el ratio está dentro del rango válido (2–184), se convierte a SpO2.

### Paso 1: importar los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ruta a un CSV de los buenos, ya recortado desde los 5 s
ruta_csv = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Procesados\raw_data_95_77_2.csv"

# Cargar CSV
df = pd.read_csv(ruta_csv)

# Extraer columnas
tiempo = df['Tiempo (ms)'].values
ir = df['IR'].values
```

### Paso 2: Eliminar la componente DC (restar la media)

```python
# Cálculo del componente DC
dc_ir = np.mean(ir)

# Señal centrada en cero (solo componente AC)
ir_ac = ir - dc_ir
```

### Paso 3: Invertir y filtrar con media móvil
La media móvil de 4 puntos se hace con un np.convolve:

```python
# Invertimos la señal
ir_invertida = -ir_ac

# Aplicamos media móvil de 4 muestras
window_size = 4
media_movil = np.convolve(ir_invertida, np.ones(window_size)/window_size, mode='same')
```

### Paso 4: Detectar valles (latidos)
Podemos usar scipy.signal.find_peaks sobre la señal invertida (los valles originales aparecen como picos).

```python
from scipy.signal import find_peaks

# Detección de picos sobre la señal suavizada
# Como es señal invertida, los picos representan los valles originales
peaks, _ = find_peaks(media_movil, distance=30)  

# Visualizar
plt.figure(figsize=(12, 5))
plt.plot(tiempo, ir, label='IR original')
plt.plot(tiempo, media_movil, label='IR invertida + filtrada', alpha=0.7)
plt.plot(tiempo[peaks], ir[peaks], 'ro', label='Valles detectados (latidos)')
plt.legend()
plt.xlabel("Tiempo (ms)")
plt.ylabel("Señal IR")
plt.title("Detección de latidos (valles)")
plt.grid(True)
plt.show()
```

```python
import os
import pandas as pd
import numpy as np

# Ruta de tus archivos
carpeta = r'C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Datos_limpios'

def estimate_spo2(ir_signal, red_signal):
    ir_signal = np.array(ir_signal, dtype=np.float64)
    red_signal = np.array(red_signal, dtype=np.float64)

    n = len(ir_signal)
    ir_mean = np.mean(ir_signal)
    red_mean = np.mean(red_signal)
    ir_ac = ir_signal - ir_mean
    red_ac = red_signal - red_mean
    ir_ac_amp = np.max(ir_ac) - np.min(ir_ac)
    red_ac_amp = np.max(red_ac) - np.min(red_ac)

    if red_mean == 0 or ir_mean == 0 or red_ac_amp == 0:
        return -999, 0

    ratio = (ir_ac_amp / ir_mean) / (red_ac_amp / red_mean)
    spo2 = 110.0 - 25.0 * ratio
    spo2 = max(0, min(100, spo2))

    return spo2, 1

# Procesar todos los CSV
for archivo in os.listdir(carpeta):
    if archivo.endswith(".csv"):
        ruta = os.path.join(carpeta, archivo)
        df = pd.read_csv(ruta)

        # Eliminar primeros 5 segundos (5000 ms)
        df = df[df['Tiempo (ms)'] >= 5000]

        # Obtener señales IR y RED
        ir = df['IR'].values
        red = df['RED'].values

        # Calcular SpO2
        spo2, valido = estimate_spo2(ir, red)

        print(f"{archivo} → SpO₂: {spo2:.2f}%" if valido else f"{archivo} → SpO₂ inválida")
```

```python
import numpy as np
import pandas as pd
import os
import glob

def calculate_spo2(ir_data, red_data):
    # Parámetros del algoritmo original
    min_autocorr_ratio = 0.5
    min_peak_height = 30
    buffer_size = len(ir_data)

    # Filtrado (sin modificaciones)
    ir_mean = np.mean(ir_data)
    red_mean = np.mean(red_data)
    ir_ac = ir_data - ir_mean
    red_ac = red_data - red_mean

    # Detección de picos
    threshold = np.max(ir_ac) * 0.5
    peaks = np.where((ir_ac[1:-1] > ir_ac[:-2]) & (ir_ac[1:-1] > ir_ac[2:]) & (ir_ac[1:-1] > threshold))[0] + 1

    if len(peaks) < 2:
        return None

    # Calcular ratios
    spo2_values = []
    for i in range(1, len(peaks)):
        start = peaks[i-1]
        end = peaks[i]
        if end > start:
            ir_segment = ir_ac[start:end]
            red_segment = red_ac[start:end]

            ir_rms = np.sqrt(np.mean(ir_segment**2))
            red_rms = np.sqrt(np.mean(red_segment**2))

            if ir_rms != 0:
                ratio = (red_rms / red_mean) / (ir_rms / ir_mean)
                spo2 = 110 - 25 * ratio
                if 0 < spo2 < 100:
                    spo2_values.append(spo2)

    if spo2_values:
        return np.mean(spo2_values)
    else:
        return None


# Ruta archivos
ruta = r"C:\Users\Elena\Desktop\GitHub\TFG-Elena-Ruiz\Datos\Procesados"
archivos = glob.glob(os.path.join(ruta, "*.csv"))

for archivo in archivos:
    try:
        df = pd.read_csv(archivo)
        ir = df['IR'].values.astype(float)
        red = df['RED'].values.astype(float)

        spo2 = calculate_spo2(ir, red)

        nombre = os.path.basename(archivo)
        if spo2 is None:
            print(f"{nombre} → SpO₂ inválida")
        else:
            print(f"{nombre} → SpO₂: {spo2:.2f}%")
    except Exception as e:
        print(f"{archivo} → Error al procesar: {e}")
```

Hay muchas señales de las que no se calcula el SPO2, esto significa que el algoritmo rechaza la señal si:

- El valor medio de la señal IR o RED es cero.

- La señal RED no tiene componente AC (es decir, no tiene pulsos medibles → red_ac_amp = 0).

Esto puede pasar si:

- La señal está completamente plana (sensor desconectado o mal colocado).

- Hay un error en el CSV (columnas vacías o mal leídas).

- El entorno era muy ruidoso y el filtro eliminó toda la señal útil.
