# Principales tipos de filtros aplicados a señales biomédicas

Este notebook tiene como objetivo representar gráficamente los tipos de filtros que principalmente se aplican a las señales biomédicas, los cuales están explicados en la memoria del proyecto.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, firwin, lfilter

# Simulación de una señal PPG contaminada con varios tipos de ruido
np.random.seed(0)
t = np.linspace(0, 10, 500)
signal_clean = 0.6 * np.sin(2 * np.pi * 1.2 * t)  # Pulso fundamental (similar a HR)
noise_high = 0.2 * np.sin(2 * np.pi * 30 * t)     # Ruido de alta frecuencia
noise_low = 0.2 * np.sin(2 * np.pi * 0.1 * t)      # Tendencia de baja frecuencia
noise = 0.15 * np.random.randn(len(t))             # Ruido blanco
ppg_noisy = signal_clean + noise_high + noise_low + noise

# --- Aplicamos diferentes tipos de filtrado ---

# Filtro paso bajo (Butterworth)
b_lp, a_lp = butter(N=2, Wn=0.2, btype='low')
ppg_lp = filtfilt(b_lp, a_lp, ppg_noisy)

# Filtro paso alto (Butterworth)
b_hp, a_hp = butter(N=2, Wn=0.5, btype='high')
ppg_hp = filtfilt(b_hp, a_hp, ppg_noisy)

# Filtro pasa banda (Butterworth)
b_bp, a_bp = butter(N=2, Wn=[0.1, 0.4], btype='band')
ppg_bp = filtfilt(b_bp, a_bp, ppg_noisy)

# Filtro FIR (Pasa bajo)
numtaps = 31
fir_coeff = firwin(numtaps, cutoff=0.2)
ppg_fir = lfilter(fir_coeff, [1.0], ppg_noisy)

# Filtro IIR (Butterworth simple)
b_iir, a_iir = butter(N=2, Wn=0.2)
ppg_iir = filtfilt(b_iir, a_iir, ppg_noisy)

# --- Representación gráfica ---
plt.figure(figsize=(16, 10))

plt.subplot(3, 2, 1)
plt.plot(t, ppg_noisy, color='grey')
plt.title('Señal PPG original con ruido')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(t, ppg_lp, color='green')
plt.title('Filtro Paso Bajo (Butterworth)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(t, ppg_hp, color='red')
plt.title('Filtro Paso Alto (Butterworth)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(t, ppg_bp, color='blue')
plt.title('Filtro Pasa Banda (Butterworth)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 2, 5)
plt.plot(t, ppg_fir, color='purple')
plt.title('Filtro FIR (Pasa Bajo)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.subplot(3, 2, 6)
plt.plot(t, ppg_iir, color='orange')
plt.title('Filtro IIR (Pasa Bajo)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

plt.tight_layout()
plt.show()
```
