```python
# Re-import required libraries and reload the Excel file after environment reset
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file again
df_final = pd.read_excel(r"C:\Users\Elena\Desktop\frecuencia_estimada.xlsx")

# Clean and structure the data
df_final = df_final.dropna().iloc[1:].reset_index(drop=True)
df_final.columns = ["Frecuencia Esperada", "Medias Móviles", "Paso Bajo"]
df_final = df_final.astype(float).round(1)

# Plot the comparison as a line chart
plt.figure(figsize=(10, 5))
plt.plot(df_final["Frecuencia Esperada"], label="Esperada", marker="o")
plt.plot(df_final["Medias Móviles"], label="Media Móvil", marker="s")
plt.plot(df_final["Paso Bajo"], label="Paso Bajo", marker="^")
plt.xlabel("Muestra")
plt.ylabel("Frecuencia cardíaca (bpm)")
plt.title("Comparación entre frecuencia estimada y esperada")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
